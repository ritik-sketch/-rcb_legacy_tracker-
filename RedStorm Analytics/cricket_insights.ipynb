```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("data/cricket_IPL.csv")


team_map = pd.read_csv("data/player_team_mapping.csv")


df.columns = df.columns.str.strip()
team_map.columns = team_map.columns.str.strip()




merged_df = df.merge(team_map[['Player_of_Match', 'Teams']], left_on='Player_Name', right_on='Player_of_Match', how='left')
merged_df.rename(columns={'Teams': 'Team_Name'}, inplace=True)
```

```python
team_centuries = merged_df.groupby('Team_Name')['Centuries'].sum().sort_values(ascending=False).head(10)

team_centuries.plot(kind='bar', color='teal', figsize=(10,6))
plt.title("Top 10 IPL Teams by Total Centuries")
plt.xlabel("Team Name")
plt.ylabel("Total Centuries")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("visuals/ipl_team_centuries.png")
plt.show()
```

```python
top_runs = merged_df.groupby('Player_Name')['Runs_Scored'].sum().sort_values(ascending=False).head(10)

top_runs.plot(kind='bar', color='green', figsize=(10,6))
plt.title("Top 10 IPL Run Scorers (20082025)")
plt.xlabel("Player Name")
plt.ylabel("Total Runs")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("visuals/top_ipl_run_scorers.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


ipl_winners = {
    'Year': [
        2008, 2009, 2010, 2011, 2012, 2013,
        2014, 2015, 2016, 2017, 2018, 2019,
        2020, 2021, 2022, 2023, 2024, 2025
    ],
    'Winner': [
        'Rajasthan Royals', 'Deccan Chargers', 'Chennai Super Kings', 'Chennai Super Kings',
        'Kolkata Knight Riders', 'Mumbai Indians', 'Kolkata Knight Riders', 'Mumbai Indians',
        'Sunrisers Hyderabad', 'Mumbai Indians', 'Chennai Super Kings', 'Mumbai Indians',
        'Mumbai Indians', 'Chennai Super Kings', 'Chennai Super Kings', 'Gujarat Titans',
        'Chennai Super Kings', 'Royal Challengers Bengaluru'
    ]
}


df_trophy = pd.DataFrame(ipl_winners)


title_counts = df_trophy['Winner'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(10,6))
title_counts.plot(kind='bar', color='gold')
plt.title(" IPL Trophy Count by Franchise (20082025)")
plt.xlabel("Team")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("visuals/ipl_trophy_count.png")
plt.show()
```
