```python
import pandas as pd
import matplotlib.pyplot as plt


rcb_runs_data = {
    'Player_Name': ['Virat Kohli', 'Philip Salt', 'Rajat Patidar', 'Jitesh Sharma', 'Devdutt Padikkal',
                    'Tim David', 'Liam Livingstone', 'Krunal Pandya', 'Mayank Agarwal', 'Romario Shepherd'],
    'Runs_Scored': [657, 403, 312, 261, 247, 187, 112, 109, 95, 70]
}

rcb_df_manual = pd.DataFrame(rcb_runs_data)

# Plot
rcb_df_manual.set_index('Player_Name').plot(kind='bar', color='crimson', figsize=(10,6))
plt.title("Top 10 RCB Run Scorers (IPL 2025)")
plt.xlabel("Player Name")
plt.ylabel("Total Runs")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("visuals/top_rcb_scorers.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

rcb_wickets_data = {
    'Player_Name': ['Yuzvendra Chahal', 'Harshal Patel', 'Mohammed Siraj', 'Vinay Kumar',
                    'Anil Kumble', 'Praveen Kumar', 'Zaheer Khan', 'Dale Steyn',
                    'Wanindu Hasaranga', 'Shane Watson'],
    'Wickets_Taken': [139, 111, 79, 72, 45, 41, 34, 30, 29, 25]
}

rcb_bowl_df = pd.DataFrame(rcb_wickets_data)

rcb_bowl_df.set_index('Player_Name').plot(kind='bar', color='darkblue', figsize=(10,6))
plt.title("Top RCB Wicket Takers (2008–2024)")
plt.xlabel("Player Name")
plt.ylabel("Total Wickets")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("visuals/top_rcb_wicket_takers.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

rcb_centuries_data = {
    'Player_Name': ['Virat Kohli', 'Chris Gayle', 'AB de Villiers', 'Manish Pandey',
                    'Devdutt Padikkal', 'Rajat Patidar', 'Cameron Green', 'Heinrich Klaasen',
                    'Will Jacks', 'Faf du Plessis'],
    'Centuries': [6, 5, 2, 1, 1, 1, 1, 1, 1, 1]
}

df_rcb_centuries = pd.DataFrame(rcb_centuries_data)

df_rcb_centuries.set_index('Player_Name').plot(kind='bar', color='orange', figsize=(10,6))
plt.title("Top 10 RCB Century Makers (Overall IPL Career)")
plt.xlabel("Player Name")
plt.ylabel("Total IPL Centuries for RCB")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("visuals/rcb_century_leaders.png")
plt.show()
```
